```python
plt.pie(df.groupby(by=["City_Category"]).size(), labels = ["A", "B", "C"], autopct = "%.2f")
plt.show()
```

```python
plt.pie(df.groupby(by=["City_Category"])["Purchase"].sum(), labels = ["A", "B", "C"], autopct = "%.2f")
plt.show()
```

```python
plt.pie(df.groupby(by=["City_Category"])["Purchase"].mean(), labels = ["A", "B", "C"], autopct = "%.2f")
plt.show()
```
